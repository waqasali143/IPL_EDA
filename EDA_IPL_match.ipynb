### Top 5 winning teams 

```python
plt.figure(figsize=(15,6))
plt.bar(list(df['WinningTeam'].value_counts()[0:5].keys()), list(df['WinningTeam'].value_counts()[0:5]))
plt.show()
```

### How many matches did each team win while batting first?

```python
df.loc[df.WonBy=='Runs', 'WinningTeam'].value_counts()[:10]
```

### Bar chart of top 5 win teams batting first 

```python
plt.figure(figsize=(15,6))
plt.bar(list(df.loc[df.WonBy=='Runs','WinningTeam'].value_counts()[0:5].keys()), 
        list(df.loc[df.WonBy=='Runs','WinningTeam'].value_counts()[0:5]), 
        color=['blue', 'orange', 'green', 'purple','yellow'])
plt.show()
```

### Top 10 Teams
Batting first win distribution in % show in Pie chart

```python
plt.figure(figsize=(7,7))
plt.title('Winning team First Batting')
plt.pie(list(df.loc[df.WonBy=='Runs','WinningTeam'].value_counts()[0:10]), 
        labels=list(df.loc[df.WonBy=='Runs','WinningTeam'].value_counts()[0:10].keys()), 
        autopct='%0.1f%%')
plt.show()
```

### How many matches did each team win while bolling first?

```python
df.loc[df.WonBy=='Wickets', 'WinningTeam'].value_counts()[:10]
```

### Draw the Pie chart for winning team with First bolling

```python
plt.figure(figsize=(7,7))
plt.title('Winning team First Bolling')
plt.pie(list(df.loc[df.WonBy=='Wickets','WinningTeam'].value_counts()[0:10]), 
        labels=list(df.loc[df.WonBy=='Wickets','WinningTeam'].value_counts()[0:10].keys()), 
        autopct='%0.1f%%')
plt.show()
```
